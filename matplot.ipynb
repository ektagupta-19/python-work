```python
# Bar Chart - Plot sales across each product category

#1. Adding labels to Axes
#2. Reducing the bar width
#3. Giving Title to the chart
#4. Modifying the ticks to show information in (million dollars)
import numpy as np
import matplotlib.pyplot as plt
lables=np.array(['Furniture', 'Technology', 'Office Supplies'])
dimension=np.array ([4110451.90, 4744557.50, 3787492.52] )

#Adding labels to Axes
plt.xlabel("Product Category", fontdict={'fontweight': 10, 'fontsize': 10, 'color': 'Blue'} )
plt.ylabel("Sales", fontdict={'fontweight': 10, 'fontsize': 10, 'color': 'Blue'} )
# plotting bar chart and setting bar width to 0.5 and aligning it to center
plt.bar(lables, dimension, width=0.5, align='center', color='cyan' ,edgecolor='orange')
plt.title("Sales for each Category\n", fontdict={'fontweight': 15, 'fontsize': 20, 'color': 'Red'})
#Modifying the ticks to show information in (million dollars)
mm= np.arange(0,6000000, 1000000)
ticks= ["{}M".format(i/1000000) for i in mm ]
plt.yticks(mm,ticks)
plt.show()
```

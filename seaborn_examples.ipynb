```python
import seaborn as sns, pandas as pd, matplotlib.pyplot as plt
```

```python
inp0 = pd.read_csv('googleplaystore_v2.csv')
```

```python
sns.displot(inp0.Rating, rug=False)
plt.show()
```

```python
sns.displot(inp0.Rating, kde=True)
plt.show()
```

```python
sns.displot(inp0.Rating, kde=False)
plt.show()
```

```python
sns.displot(inp0.Rating,)
plt.show()
```

```python
inp0["Android Ver"].value_counts().plot.barh()
plt.show()
```

```python
sns.boxplot(inp0.Rating)
plt.show()
```

```python
#inp0.head()
inp0.Genres.value_counts().head(4)

inp1=inp0[inp0.Genres.isin(['Tools','Entertainment','Education','Medical'])]
#inp1
sns.boxplot(inp1.Genres, inp1.Rating)

plt.show()
```
